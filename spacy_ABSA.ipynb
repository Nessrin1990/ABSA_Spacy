```python
import pandas as pd

dataset = pd.read_csv('resturant.csv' ,encoding = "utf-8")

dataset.head(5)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

dataset.text = dataset.text.str.lower()

aspect_terms = []
for text in nlp.pipe(dataset.text):
    chunks = [(chunk.root.text) for chunk in text.noun_chunks if chunk.root.pos_ == 'NOUN']
    aspect_terms.append(' '.join(chunks))
dataset['aspect_terms'] = aspect_terms
dataset.head(10)
```

```python
sentiment_terms = []
for review in nlp.pipe(dataset['text']):
        if review.is_parsed:
            sentiment_terms.append(' '.join([token.lemma_ for token in review if (not token.is_stop and not token.is_punct and (token.pos_ == "ADJ" or token.pos_ == "VERB"))]))
        else:
            sentiment_terms.append('')  
dataset['sentiment_terms'] = sentiment_terms
dataset.head(10)
```

```python
sentences = dataset['text'].head(10)
```

```python
sentences = [
  'The food we had yesterday was delicious',
  'My time in Italy was very enjoyable',
  'I found the meal to be tasty',
  'The internet was slow.',
  'Our experience was suboptimal'
]
```

```python
for sentence in sentences:
  doc = nlp(sentence)
  for token in doc:
    print(token.text, token.dep_, token.head.text, token.head.pos_,
      token.pos_,[child for child in token.children])
```

```python
for sentence in sentences:
  doc = nlp(sentence)
  descriptive_term = ''
  for token in doc:
    if token.pos_ == 'ADJ':
      descriptive_term = token
  print(sentence)
  print(descriptive_term)
```

```python
for sentence in sentences:
  doc = nlp(sentence)
  descriptive_term = ''
  for token in doc:
    if token.pos_ == 'ADJ':
      prepend = ''
      for child in token.children:
        if child.pos_ != 'ADV':
          continue
        prepend += child.text + ' '
      descriptive_term = prepend + token.text
  print(sentence)
  print(descriptive_term)
```

```python
aspects = []
for sentence in sentences:
  doc = nlp(sentence)
  descriptive_term = ''
  target = ''
  for token in doc:
    if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
      target = token.text
    if token.pos_ == 'ADJ':
      prepend = ''
      for child in token.children:
        if child.pos_ != 'ADV':
          continue
        prepend += child.text + ' '
      descriptive_term = prepend + token.text
  aspects.append({'aspect': target,
    'description': descriptive_term})
print(aspects)
```

```python
from textblob import TextBlob
for aspect in aspects:
  aspect['sentiment'] = TextBlob(aspect['description']).sentiment
print(aspects)
```

```python
from textblob.classifiers import NaiveBayesClassifier
import nltk
nltk.download('punkt')
# We train the NaivesBayesClassifier
train = [
  ('Bad girl.', 'negative'),
  ('amazing  place', 'positive'),
  ('fast service', 'positive'),
  ('Very expensive price', 'negative'),
  ('delicious food.', 'neg')
]
cl = NaiveBayesClassifier(train)
# And then we try to classify some sample sentences.
blob = TextBlob("Bad girl. amazing  place. Very expensive price.", classifier=cl)
for s in blob.sentences:
  print(s)
  print(s.classify())
```
